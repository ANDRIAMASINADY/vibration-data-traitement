```python
import numpy as np
from numpy.fft import fft, fftfreq
import pandas as pd
import matplotlib.pyplot as plt

# read csv file and stored it in s

s = pd.read_csv("C:/Users/Angeleco/Desktop/P4 [517]/ISO TEMPOREL Y [5235]/2021_9_15/ISO TEMPOREL Y.csv", sep = ";" , skiprows = 1)
   
t=s["time[ms]"]
y=s[" amplitude[mm/s]"]

N=y.size # Nombre d'echantillon
print(N)
Duration =8 # La longueur du signaltemporel en seconde(Durée)
Te = 1/4096 # Temps d'échantillonage

X = fft(y)  # Transformée de fourier

freq = fftfreq(y.size, 1/4096)  # Fréquences de la transformée de Fourier

# On prend la valeur absolue de l'amplitude uniquement pour les fréquences positives et normalisation

X_abs = np.abs(X[:N//2])*2.0/N

# Creation du graph FFT et On garde uniquement les fréquences positives

fig = plt.figure(1)
freq_pos = freq[:N//2]
plt.plot(freq_pos, X_abs, label="Amplitude absolue")
plt.xlim(0, 1600)  # On réduit la plage des fréquences à la zone utile
plt.grid()
plt.xlabel("Fréquence (Hz)")
plt.ylabel("Amplitude |X(f)| [mm/s]_O-P")
plt.title("Transformée de Fourier du 2021_9_15 of Signal waveform  b_2021 ")
plt.show()
fig.savefig("Spectre FFT")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

#creat a list for existing folder

data_list = [ "2020_9_25", "2021_9_15", "2021_9_15_1"]

v = ["a_2020", "b_2021","c_2021"] # legend list

# read csv file and stored it in vibration_data
for i in data_list: 
    vibration_data = pd.read_csv("C:/Users/Angeleco/Desktop/P4 [517]/ISO TEMPOREL Y [5235]/"+i+"/ISO TEMPOREL Y.csv", sep = ";" , skiprows = 1)
# create graph waveform    
    fig = plt.figure(0)
    x=vibration_data["time[ms]"]
    y=vibration_data[" amplitude[mm/s]"]
    plt.plot(x,y,label=v[data_list.index(i)])
    plt.xlabel("Time[ms]")
    plt.ylabel("Amplitude[mm/s]")
    plt.title("Signal waveform comparaison")
    plt.legend(loc ="right")
    fig.savefig("Signal wave form.png")
    
plt.show()
```

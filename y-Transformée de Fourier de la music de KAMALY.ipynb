```python
from numpy.fft import fft, fftfreq
import scipy.io.wavfile as wavfile
import matplotlib.pyplot as plt
import numpy as np

# Calcul FFT
rate, data = wavfile.read('KAMALY.wav') 
x = data[:, 0]
print(x)
N = x.size
print(N)
X = fft(x)  # Transformée de fourier
freq = fftfreq(x.size, d=1/rate)  # Fréquences de la transformée de Fourier

# On prend la valeur absolue de l'amplitude uniquement pour les fréquences positives et normalisation

X_abs = np.abs(X[:N//2])*2.0/N
fig = plt.figure(2)
freq_pos = freq[:N//2]
plt.plot(freq_pos, X_abs, label="Amplitude absolue")
plt.xlim(0, 2000)  # On réduit la plage des fréquences à la zone utile
plt.grid()
plt.xlabel("Fréquence (Hz)")
plt.ylabel("Amplitude |X(f)|")
plt.title("Transformée de Fourier de la music de KAMALY ")
plt.show()
fig.savefig("Spectre FFT")
```
